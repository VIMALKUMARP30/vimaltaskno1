```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.DataFrame({
 'humidity':[65,70,80,60,75,85,90,55],
 'pressure':[1012,1010,1008,1015,1011,1007,1005,1016],
 'temperature':[30,28,26,32,29,25,24,33]
})

df
```

```python
X = df[['humidity','pressure']]
y = df['temperature']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
result = X_test.copy()
result["Actual"] = y_test.values
result["Predicted"] = y_pred
result
```

```python
x = range(len(y_test))

plt.plot(x, y_test.values, marker='o', label="Actual")
plt.plot(x, y_pred, marker='s', label="Predicted")

plt.xlabel("Sample Index")
plt.ylabel("Temperature (°C)")
plt.title("Actual vs Predicted Temperature")
plt.legend()
plt.grid(True)
plt.show()
```
